```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('/content/drive/MyDrive/20250619_112801 (1).jpg',target_size=(200,200))
```

```python
 img
```

```python
type(img)
```

```python
datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
i=0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/content/drive/MyDrive/deep learning'):
  i=i+1

  if i==10:
    break
```
